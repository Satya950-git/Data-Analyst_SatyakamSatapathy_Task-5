```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Load the dataset
df = pd.read_csv('Downloads/student_dataset.csv')

# 1. Data Exploration

# Check for missing values
print("Missing Values:\n", df.isnull().sum())

# Boxplots for outliers
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x=df['Study Hours'])
plt.title('Boxplot of Study Hours')
plt.subplot(1,2,2)
sns.boxplot(x=df['Attendance'])
plt.title('Boxplot of Attendance')
plt.tight_layout()
plt.show()

# Pairplot to visualize relationships and trends
sns.pairplot(df, hue='Pass', palette='coolwarm', diag_kind='kde')
plt.show()

# 2. Model Training

# Features and target
X = df[['Study Hours', 'Attendance']]
y = df['Pass']

# Train Logistic Regression model
logreg = LogisticRegression()
logreg.fit(X, y)

# 3. Model Evaluation

# Predictions
y_pred = logreg.predict(X)

# Accuracy
accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix
conf_matrix = confusion_matrix(y, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```
